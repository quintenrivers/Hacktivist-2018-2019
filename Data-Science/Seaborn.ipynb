## Intro to Seaborn
Seaborn is a data visualization library built on top of Matplotlib. It has many of the same functionalities as Matplotlib while also being significantly easier to use.
## Installation and Usage

````python
import seaborn as sns
%matplotlib inline
````

# Plotting

## Categorical Plotting

````python
tips = sns.load_dataset('tips') # tips is a seaborn built-in dataframe
tips.head() # head finds the front of the dataframe
````

#### Barplot
Standard bar graph

````python
sns.barplot(x='sex',y='total_bill',data=tips)
# data is a dataframe
# x and y are column names in the dataframe
````

#### Countplot
Similar to barplot but counts the number of instances

````python
sns.countplot(x='sex',data=tips)
````

#### Boxplot

````python
sns.boxplot(x="day",
            y="total_bill",
            data=tips,
            palette='rainbow')
````

````python
sns.boxplot(x="day",
            y="total_bill",
            hue="smoker", # hue gives a secondary category to divide the data into
            data=tips,
            palette="coolwarm")
````

There are many more types of categorical plots but these are the most used

## Distribution Plots

#### Distplot

````python
sns.distplot(tips['total_bill'],bins=30) # bins is a way of grouping data
````

#### Jointplot
Matches up distplots with bivariate (two variable) data
> The kind can be scatter, hex, reg, resid, kde

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
````

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
````

````python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
````

#### Pairplot
Plots pairwise relationships within a dataframe. This is mostly useful for finding correlations you may not have noticed on your own

````python
sns.pairplot(tips) # the diagonals show distributions
````

````python
sns.pairplot(tips,hue='sex',palette='coolwarm')
````

#### Regression Plot
Plots linear regression. Useful for general linear regression but not necessarily for machine learning purposes

````python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
````
